```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/1) iris.csv')
df.head()
```

```python
#Summary Statistics
print("Shape:", df.shape)
print("\nDescribe:")
df.describe()
# Mean, Median, Mode, Std
for col in df.select_dtypes(include=np.number).columns:
    print(f"\n{col}:")
    print(f"  Mean: {df[col].mean():.2f}")
    print(f"  Median: {df[col].median():.2f}")
    print(f"  Mode: {df[col].mode()[0]:.2f}")
    print(f"  Std Dev: {df[col].std():.2f}")
```

```python
# Histograms
df.hist(figsize=(10, 6), bins=20, color='steelblue', edgecolor='black')
plt.suptitle('Feature Distributions - Iris Dataset', fontsize=14)
plt.tight_layout()
plt.savefig('histograms.png', dpi=150)
plt.show()
```

```python
#Boxplots
num_cols = df.select_dtypes(include=np.number).columns

fig, axes = plt.subplots(1, len(num_cols), figsize=(14, 5))
for i, col in enumerate(num_cols):
    sns.boxplot(y=df[col], ax=axes[i], color='orange')
    axes[i].set_title(col)
plt.suptitle('Boxplots - Iris Dataset')
plt.tight_layout()
plt.savefig('boxplots.png', dpi=150)
plt.show()
```

```python
#Scatter Plot
sns.pairplot(df, hue='species', palette='Set2')
plt.suptitle('Pairplot - Iris Dataset', y=1.02)
plt.savefig('scatterplot.png', dpi=150)
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.select_dtypes(include=np.number).corr(),
            annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap - Iris Dataset')
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=150)
plt.show()
```
